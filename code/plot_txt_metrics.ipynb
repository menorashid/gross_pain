```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
subject_dict = {'as' : 'aslan',
                'br' : 'brava',
                'he' : 'herrera',
                'in' : 'inkasso',
                'ju' : 'julia',
                'ka' : 'kastanjett',
                'na' : 'naughty_but_nice',
                'si' : 'sir_holger'}
```

```python
results_folder = '../data/output/bg_50epochs_march31st/'
loss_file = 'debug_log_training.txt'

loss_df = pd.DataFrame()
counter = 0
for root, dirs, files in os.walk(results_folder):
    for file in files:
        if file == loss_file:
            subject = subject_dict[root[-17:-15]]
            df = pd.read_csv(root + '/' + file, index_col = None, header=0)
            if counter == 0:
                loss_df.insert(counter, column='iterations', value=df['#iteration'])
            loss_df.insert(counter, column=subject, value=df['loss'])
            counter+=1

            
loss_df.set_index('iterations', inplace=True)
print(loss_df.head())
print(len(loss_df))
```

```python
plt.figure()
loss_df.plot(figsize=(20,20))
```

```python
results_folder = '../output/'
loss_file = 'debug_log_testing.txt'

loss_df = pd.DataFrame()
acc_df = pd.DataFrame()
counter = 0
for root, dirs, files in os.walk(results_folder):
    for file in files:
        if file == loss_file:
            subject = subject_dict[root[-31:-29]]
            df = pd.read_csv(root + '/' + file, index_col = None, header=0)
            df.columns = ['iterations', 'acc', 'loss']
            print(df.head())
            if counter == 0:
                loss_df.insert(counter, column='iterations', value=df['iterations'])
            loss_df.insert(counter, column=subject, value=df['loss'])
            acc_df.insert(counter, column=subject, value=df['acc'])
            counter+=1

            
loss_df.set_index('iterations', inplace=True)
print(loss_df.head())
print(len(loss_df))

plt.figure()
loss_df.plot(figsize=(20,20))
```

```python
plt.figure()
acc_df.plot(figsize=(20,20))
```
